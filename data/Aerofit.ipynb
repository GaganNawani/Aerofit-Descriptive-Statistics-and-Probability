## 3.2 For categorical variable(s): Boxplot (10 Points)

```python
columns = ['Age','Education','Usage','Fitness','Income','Miles']

plt.figure(figsize = (18,12)).suptitle('BiVariate Analysis', fontsize = 22)
count = 0
for i in range(6):
    plt.subplot(2,3,i+1)
    sns.boxplot(data = df, x = 'Product', y = columns[count], palette = 'Set1')
    plt.title(f"Product vs {columns[count]}")
    count+=1
```

## Observations
Product vs Age:
1. Same Age median values for customers who bought KP281 and KP481 products.
2. Customers whose age lies between 25-30 are more likely to buy KP781 product.
3. We can also see that there are a few outlier values in KP781, because 90-95% of customers with age 25-40 have bought KP781 and the few people in the age range of 40-50 have also bought the KP781 product.

Product vs Education:
1. Customers whose Education is more than 16 have higher chance of buying the KP781 product.
2. Customers whose Education is in the range of 14-16 tends to buy either KP281 or KP481 product.
3. We do not see any outliers in this plot.

Product vs Usage:
1. We can see people with usage of 3-4 days a week buys more KP281 products.
2. People with usage of 4-5 days a week buys KP781 products.
3. We can also observe that 90-95% KP481 is bought by people who workout for 3 days a week with a few outliers of 4 and 5 days week.

Product vs Fitness:
1. The customers who are fitter i.e. whose fitness is more than 3 tends to buy the KP781 products while the rest tends to buy the KP 281 or KP481 products.
2. We can see a few outliers for KP281 and KP481 because most of the customers buying these 2 products have the fitness level of 3 and very few of the customers other fitness levels.
4. We can see that the products KP281 and KP481 have a few outliers in this plot.

Product vs Income:
1. We can clearly see that the customers with higher income i.e. from 60k - 100k buy the KP781 product.
2. Most of the customers with 50k income buy the KP281 and KP481 product.

Product vs Miles:
1. The customers who expects to walk/run more than 150 miles buy the KP781 product.
2. The customers who expects to walk/run close to 90-100 miles usually buy the KP281 or KP481 product.
3. We can also see there are a few outliers for the KP281 and KP481 product in this plot.

## 3.3 For correlation: Heatmaps, Pairplots(10 Points)

```python
# Converting the categorical features into numerical features

new_df = df.copy()

new_df['Product'] = new_df['Product'].replace(('KP281','KP481','KP781'),(1,2,3))
new_df['Gender'] = new_df['Gender'].replace(('Male','Female'),(0,1))
new_df['MaritalStatus'] = new_df['MaritalStatus'].replace(('Single','Partnered'),(0,1))
```

```python
numerical_var = new_df[['Product','Gender','MaritalStatus','Age','Education','Usage','Fitness','Income','Miles']]

sns.heatmap(data = numerical_var.corr(),cmap = 'Blues', annot = True)
plt.title('Heatmap')
plt.show()
```

```python
sns.pairplot(data = df, hue = 'Product')
plt.show()
```

## Observations:
1. We can see that the Usage and Fitness are slightly correlated. Because more usage can mean better fitness.
2. We can see that the Product and Income is also decently correlated. Because customers with decent income only can afford the treadmills.
3. Usage and Miles are also correlated because customers who will workout atleast 3-5 times a day will have the higher avg miles.
4. Miles and Fitness are also strongly correlated because the higher the avg miles, the better the fitness of the customers will be.
5. From the above heatmap we can see that the Fitness and Products are decently correlated because people who have rated themselves to be fit will be more into buying the treadmills.
